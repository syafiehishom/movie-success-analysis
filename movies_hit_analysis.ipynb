```python
# Data Cleaning
```

```python
# Removing Unnecessary Columns
```

```python
df = df.drop(columns=["tagline", "overview", "homepage", "keywords"])
```

```python
# Remove missing critical values
```

```python
df = df.dropna(subset=["runtime", "release_date"])
```

```python
# Keep Only Relevant Columns
```

```python
columns_to_keep = [
    "title",
    "revenue",
    "budget",
    "genres",
    "vote_average",
    "release_date"
]

df = df[columns_to_keep]
```

```python
# Feature Engineering
```

```python
# Extract Genres from JSON
```

```python
def extract_genres(x):
    try:
        genres = json.loads(x)
        return [i['name'] for i in genres]
    except:
        return []

df['genres'] = df['genres'].apply(extract_genres)
```

```python
# Create main genre
```

```python
df['main_genre'] = df['genres'].apply(lambda x: x[0] if len(x) > 0 else None)
```

```python
# Convert date format
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
# Remove rows with missing main genre
```

```python
df = df.dropna(subset=['main_genre'])
```

```python
# Exploratory Data Analysis (EDA)
```

```python
# Check relationship: Budget vs Revenue
```

```python
df[['budget', 'revenue']].corr()
```

```python
# Visual: Budget vs Revenue
```

```python
plt.scatter(df['budget'], df['revenue'], alpha=0.5)
plt.xlabel('budget')
plt.ylabel('revenue')
plt.title('budget vs revenue')
plt.show()
```

```python
# Top 10 Highest Revenue Movies
```

```python
df.sort_values(by='revenue', ascending=False).head(10)
```

```python
# Genre Analysis (Average Revenue)
```

```python
df.groupby('main_genre')['revenue'].mean().sort_values(ascending=False)
```

```python
# Visualization: Revenue by Genre
```

```python
df.groupby('main_genre')['revenue'].mean()\
.sort_values(ascending=False)\
.plot(kind='bar')

plt.title("Average Revenue by Genre")
plt.xlabel("Genre")
plt.ylabel("Average Revenue")
plt.show
```
