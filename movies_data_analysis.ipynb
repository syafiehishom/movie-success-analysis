```python
## Data Cleaning
## Removing unnecessary columns
```

```python
df = df.drop(columns=["homepage", "tagline", "overview", "keywords"])
```

```python
df.head()
```

```python
df = df.dropna(subset=["runtime", "release_date"])
```

```python
columns_to_keep = [
    "title", 
    "revenue",
    "budget",
    "genres",
    "vote_average",
    "release_date"
]
```

```python
df = df[columns_to_keep]
```

```python
df.head(10)
```

```python
import json

def extract_genres(x):
    try:
        genres = json.loads(x)
        return[i['name'] for i in genres]
    except:
        return[]

df['genres'] = df['genres'].apply(extract_genres)
```

```python
df.head(10)
```

```python
df['main_genre'] = df['genres'].apply(lambda x: x[0] if len(x) > 0 else None)
```

```python
df.head(10)
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
df.info()
```

```python
df = df.dropna(subset=['main_genre'])
```

```python
df.info()
```

```python
#Do bigger budgets = Higher revenue?
```

```python
df[['budget', 'revenue']].corr()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['budget'], df['revenue'], alpha=0.5)
plt.xlabel('budget')
plt.ylabel('revenue')
plt.title('budget vs revenue')
plt.show()
```

```python
df.sort_values(by='revenue', ascending=False).head(10)
```

```python
#Genre Analysis
df.groupby('main_genre')['revenue'].mean().sort_values(ascending=False)
```

```python
# Analysis: Highest Average Revenue by Genre
df.groupby('main_genre')['revenue'].mean()\
    .sort_values(ascending=False)\
    .plot(kind='bar')
plt.title('Average Revenue by Genre')
plt.show()
```

```python
## Conclusion

- Budget has a strong positive relationship with revenue
- Animation and Adventure are top-performing genres
- Some low-budget films achieve high revenue, indicating other factors matter
```
